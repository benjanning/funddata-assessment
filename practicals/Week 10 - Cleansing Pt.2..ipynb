Adapt the code to capitalise the first letter of the iris species, using regular expressions.

```python
import re
```

```python
# https://stackoverflow.com/a/1393367

import urllib.request

url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

iris
```

```python
strip_iris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-z]+)')
```

```python
[strip_iris.sub(r'\5,\4,\3,\2,\1', line) for line in iris if line]
```

```python
re.sub('setosa', 'Setosa', iris)
```

```python
# Finding the Words [1]

species = re.search('(\w+),(\w+),(\w+)', 'setosa,versicolor,virginica')
species.groups()
```

```python
newspecies = re.sub(species, 'Setosa, Versicolor, Virginica')
```

```python
species = 'setosa, versicolor, virginica'

newspecies = re.sub(r'\Setosa, Versiclor, Virginica', 'robert', species)
newspecies
```
