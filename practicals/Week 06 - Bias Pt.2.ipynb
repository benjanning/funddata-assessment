## Give Three Real-World Examples of Different Types of Cognitive Bias.

### Bandwagon Effect:
The bandwagon effect is when people start doing something because everybody else seems to be doing it.

Examples include:  

People voting for the same political party because they already seem to have a lot of support.  People want to be a part of the majority.

Consumers buying the same expensive products in order to achieve status.

During the dotcom bubble of the late 1990's people invested in projects that had little to no business model just because they were part of a euphoric trend.  A parallel can be drawn to investing in Cryptocurrencies today.  A handful of projects have fundamentals and some history, others have none at all and investors put in their money because others have achieved wealth through the space.  


### Framing Effect:
The framing effect refers to how a person may react depend on how information is presented to them.

Examples include:

A doctor telling a patient that they have a 20% chance of dying, or telling them they have a 80% chance of survival.

The story of Tom Sawyer, where Tom positively frames the task of whitewashing fences and gets his friends to pay him for the "privilege" of doing the work.

### Availability Bias, or Recency Bias:

Recency Bias refers to people remembering past events more clearly and perceiving the chances of something happening to be greater than they actually are, based on recent events.

Examples include:

Shark attacks. If there were no shark attacks for 20 years, but one shark attack makes headlines, people will be inclined to stay out of the water and completely overestimate the chances of shark attacks.

In the stock market, the number of participants who enter when the price is in a "bubble" is largest.  This is due to the media coverage and people perceiving the price to only go up (based on recency).

Calculating the standard deviation of the following two lists of numbers

```python
# Numerical arrays.
import numpy as np

# Plots.
import matplotlib.pyplot as plt
```

## List 1

```python
# A list of numbers - four small and one big.
numbers1 = np.array([1, 1, 1, 1, 10])
```

```python
# Their mean.
np.mean(numbers1)
```

```python
# Generate a sample of values - note we can see the population standard deviation.
x = np.random.normal(10.0, 1.0, 1000)
x
```

```python
# A list of numbers - all close to each other.
numbers2 = np.array([2, 2, 3, 3, 4])

# Their mean.
np.mean(numbers2)
```

```python
# Calculate the mean.
x_mean = x.mean()

# Subtract the mean from each of the values.
zeroed = x - x_mean

# Have a look at the zeroed values.
zeroed
```

```python
zeroed.mean()
```

```python
# This will give us a better view of it - correct to four decimal places.
print(f'{zeroed.mean():0.4f}')
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), zeroed, 'k.')

# Plot the y=0 line.
ax.axhline(y=0.0, color='grey', linestyle='-');
```

```python
# Sum the array.
print(f'{zeroed.sum():0.4f}')
```

```python
# Average absolute value.
np.mean(np.abs(zeroed))
```

```python
# Square the values.
np.square(zeroed)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the squared zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), np.square(zeroed), color='green', marker='.', linestyle='none')

# Plot the zeroed array, each value spaced out evenly along the x axis.
# Note the x axis is just the position of the value in the zeroed array.
ax.plot(range(len(zeroed)), zeroed, 'k.')

# Plot the y=0 line.
ax.axhline(y=0.0, color='grey', linestyle='-');
```

```python
# Calculate the average squared result.
np.mean(np.square(zeroed))
```

```python
# Calculate the square root of the average squared result.
np.sqrt(np.mean(np.square(zeroed)))
```

Compare Results of the built in function to the full calculation 

```python
# The full calculation using the original array.
np.sqrt(np.mean(np.square(x - np.mean(x))))
```

## Exercise - Bessel's Correction
Show that the difference between the standard deviation calculations is greatest for small sample sizes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Calculate the mean of the first array using Numpy

array1 = np.array([1,1,1,2,3])
np.mean(array1)
```

```python
# Calculate the mean of the second array using Numpy

array2 = np.array([1,5,7,3,10])
np.mean(array2)
```

```python
# Get the zeroed mean of the first array

mean1 = array1.mean()

zeroedm = array1 - mean1

zeroedm.mean()

print (f'{zeroedm.mean():0.4f}')
```

```python
# Square the zeroed mean

np.square(zeroedm)
```

```python
## Get the zeroed mean of the second array

mean2 = array2.mean()

zeroedm2 = array2 - mean2

zeroedm2.mean()

print (f'{zeroedm2.mean():0.4f}')
```

```python
np.square(zeroedm2)
```

```python
# Calculate the standard deviation of the first array

zeroedm.std()
```

```python
# Calculate the standard deviation of the second array

zeroedm2.std()
```
