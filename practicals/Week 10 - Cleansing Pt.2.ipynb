Adapt the code to capitalise the first letter of the Iris species, using regular expressions.

```python
import re
import urllib.request

# URL of the iris dataset
url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Read the dataset from the URL
iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

# Compile a regular expression to extract the iris species
strip_iris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-z]+)')

# Use re.sub() to capitalize the first letter of the iris species. [3]
modified_iris = [strip_iris.sub(lambda x: f"{x.group(5).capitalize()},{x.group(4)},{x.group(3)},{x.group(2)},{x.group(1)}", line) for line in iris if line]

# Make the results more easy to read using pprint [2]
import pprint

pprint.pprint(modified_iris)
```
